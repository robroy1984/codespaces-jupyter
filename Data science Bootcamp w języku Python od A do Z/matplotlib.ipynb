```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0,1],[0,3])
```

```python
_ = plt.plot([0,1,3],[0,3,4], label='line',color='green')
plt.legend() #wyswietlanie legendy
plt.xlabel("oś x")
plt.ylabel("oś y")
plt.title("Wykres")
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos =np.cos(x)

plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

<h1> Tworzenie wielu wykresów

```python
fig = plt.figure(figsize=(8,8)) #kontrolowanie rozmiaru wykresów
plt.subplot(131) #121 - 1-liczba wierszy w siatce wykresów, 3-liczba kolumn, 1-odnosimy sie do konretnego wykresu
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[5,2,3], 'r--') #r-- linia czerwona. przerywana
plt.subplot(133)
plt.plot([0,1,2],[5,2,3], 'ro') #ro - punktowy
```

<h1>Wykresy słupkowe

```python
plt.bar(x=[0,1,2,3], height=[1.2,3.2,1.2,5.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2,3.2,1.2,5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2,3.2,1.2,5.2])
```

<h1>Wykresy punktowe

```python
x3=np.random.randn(300)*50
plt.scatter(x=x1,y=x2, s=x3, alpha=.5)
```
