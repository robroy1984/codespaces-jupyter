<h1> Obliczanie nowych kolumn

```python
df = fetch_financial_data('UBER')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Average'] = (df.Open + df.Close) / 2
```

```python
df = df.sort_index()
```

```python
df
```

```python
df['pct_dynamic_close'] = df.Close / df.Close.shift(1) - 1
df['Daily_change'] = df['pct_dynamic_close']
df
```

```python
df.Daily_change.min()
```

```python
df.Daily_change.max()
```

```python
df.Daily_change.hist(bins=100)
```

<h1>Filtrowanie danych

```python
df_positive = df[df.Daily_change > 0]
df_positive
```

```python
#dodatnia stopa zwrotu
df_positive.Daily_change.mean()
```

```python
df_negative = df[df.Daily_change < 0]
df_negative
```

```python
#ujemna stopa zwrotu
df_negative.Daily_change.mean()
```
