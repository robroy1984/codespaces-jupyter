```python
import pandas as pd
import numpy as np
import pandas_datareader as web
```

```python
def fetch_financial_data(company='AMZN',data_source='stooq'):
    return web.DataReader(company,data_source)
```

```python
dict = {'AAPL':'apple',
        'AMZN':'amazon',
        'GOOGL':'google',
        'UBER':'uber'

}
for company in dict.keys():
    globals()[company] = fetch_financial_data(company)
    globals()[company].columns = [company + "_" + globals()[company].columns]
    
```

```python
dict.values()
```

```python
#konkatencja tabel
df = pd.concat(objs=[AAPL,AMZN,GOOGL,UBER], axis=1)
```
